Importação de bilbiotecas

```python
import pandas as pd
from collections import Counter
from datetime import date, datetime
```

Carregando o Excel (era pra ser csv, mas os dados não vieram no formato)

```python
dataset_csv = pd.read_excel('/content/sample_data/dados-Gov.xlsx')
```

Renomeando coluna de ID

```python
rj_gov_dt = dataset_csv.rename(columns={'ÿid': 'id'})
```

Verificando valores não nulos na coluna cbo

```python
rj_gov_dt.loc[pd.notnull(rj_gov_dt['cbo'])].head(5)
```

Verificando qual a maior ocorrencia na coluna cbo

```python
cbo_counter = rj_gov_dt['cbo']

cbo_counter_gov = Counter(cbo_counter)

cbo_counter_gov.most_common()
```

Substituindo valores nulos por valores padrão

```python
rj_gov_dt['cbo'].fillna('3222 - Técnicos e auxiliares de enfermagem', inplace=True)

rj_gov_dt.head(10)
```

Pesquisando maiores valores recorrentes em paisOrigem

```python
gov_counter = Counter(rj_gov_dt['paisOrigem'])

gov_counter.most_common()
```

Substituindo alores undefined por Brasil

```python
rj_gov_dt['paisOrigem']= rj_gov_dt['paisOrigem'].replace(['undefined'],'Brasil')
```

Buscando a contagem de valoes nulos

```python
rj_gov_dt.isnull().sum()
```

Buscando valores constantes em dataEncerramento

```python
data_counter = Counter(rj_gov_dt['dataEncerramento'])

data_counter.most_common()
```

Inserindo o valor mais recorrente na coluna dataEncerramento

```python
rj_gov_dt['dataEncerramento'].fillna('2021-05-19T03:00:00.000Z', inplace=True)
```

Formatando os valores de data na coluna

```python
rj_gov_dt['dataEncerramento']= pd.to_datetime(rj_gov_dt['dataEncerramento']).dt.tz_localize(None) 

rj_gov_dt['dataEncerramento'] = pd.to_datetime(rj_gov_dt['dataEncerramento'], format='%Y%m%d')
rj_gov_dt['dataEncerramento'] = pd.to_datetime(rj_gov_dt["dataEncerramento"].dt.strftime('%d-%m-%Y'))
```

Valores diferentes do mais recorrente

```python
rj_gov_dt.loc[rj_gov_dt['dataEncerramento'] != '2021-05-19']
```
