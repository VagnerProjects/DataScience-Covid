```python
import pandas as pd
from collections import Counter
from datetime import date, datetime
```

```python
dataset_csv = pd.read_excel('/content/sample_data/dados-Gov.xlsx')
```

```python
rj_gov_dt = dataset_csv.rename(columns={'ÿid': 'id'})
```

```python
rj_gov_dt.loc[pd.notnull(rj_gov_dt['cbo'])]
```

```python
cbo_counter = rj_gov_dt['cbo']

cbo_counter_gov = Counter(cbo_counter)

cbo_counter_gov.most_common()

rj_gov_dt['cbo'].fillna('3222 - Técnicos e auxiliares de enfermagem', inplace=True)

rj_gov_dt

```

```python

gov_counter = Counter(rj_gov_dt['paisOrigem'])

gov_counter.most_common()
```

```python

rj_gov_dt['paisOrigem']= rj_gov_dt['paisOrigem'].replace(['undefined'],'Brasil')

```

```python
rj_gov_dt.isnull().sum()
```

```python
#data = datetime.strptime('26/08/2018', '%d/%m/%Y').date()

data_test = Counter(rj_gov_dt['dataEncerramento'])

data_test.most_common()

rj_gov_dt['dataEncerramento'].fillna('2021-05-19T03:00:00.000Z', inplace=True)
```

```python

rj_gov_dt['dataEncerramento']= pd.to_datetime(rj_gov_dt['dataEncerramento']).dt.tz_localize(None) 

rj_gov_dt['dataEncerramento'] = pd.to_datetime(rj_gov_dt['dataEncerramento'], format='%Y%m%d')
rj_gov_dt['dataEncerramento'] = pd.to_datetime(rj_gov_dt["dataEncerramento"].dt.strftime('%d-%m-%Y'))

rj_gov_dt.loc[rj_gov_dt['dataEncerramento'] != '2021-05-19']
```
